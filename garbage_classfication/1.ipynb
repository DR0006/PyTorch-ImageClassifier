```python
import matplotlib.pyplot as plt

# 定义类别名称的中文对应关系
class_mapping = {
    "battery": "电池",
    "biological": "生物制品",
    "brown-glass": "棕瓶",
    "cardboard": "纸板",
    "clothes": "衣物",
    "green-glass": "绿瓶",
    "metal": "金属",
    "paper": "纸类",
    "plastic": "塑料",
    "shoes": "鞋子",
    "trash": "其他垃圾",
    "white-glass": "白瓶",
}

new_categories = list(class_mapping.values())
new_colors = ['skyblue', 'lightgreen', 'lightcoral', 'lightsalmon', 'lightseagreen', 'plum',
              'cyan', 'lime', 'tomato', 'coral', 'mediumseagreen', 'violet']

counts = [945, 985, 607, 891, 5325, 629, 769, 1050, 865, 1977, 697, 775]

# 创建一个新的图形
fig, ax = plt.subplots()

# 绘制表格
table_data = [(category, count) for category, count in zip(new_categories, counts)]
table = ax.table(cellText=table_data, colLabels=['类别', '数量'], loc='center', cellLoc='center', colColours=new_colors)

# 调整表格布局
table.auto_set_font_size(False)
table.set_fontsize(10)
table.scale(1, 1.5)  # 调整表格的大小

# 隐藏坐标轴
ax.axis('off')

plt.show()
```
