```python
import matplotlib.pyplot as plt

# Accuracy data
accuracies = [91.50, 92.45, 92.50, 94.50, 95.50, 95.54, 96.50, 96.89, 97.74]

# Corresponding run numbers
runs = list(range(1, len(accuracies) + 1))

# Plot the line chart
plt.plot(runs, accuracies, marker='o', linestyle='-', color='b')

# Set the chart title and labels
plt.title('Accuracy Variation with Number of Runs')
plt.xlabel('Number of Runs')
plt.ylabel('Accuracy (%)')

# Display the legend in English
plt.legend(['Accuracy'], loc='lower right')

# Save the plot as a PNG file
plt.savefig('accuracy_plot.png')

# Display the chart
plt.show()
```
